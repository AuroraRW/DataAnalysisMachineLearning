## Dictionaries

```python
student = {'name':'Tom', 'ID': 123}
```

```python
student['name']
```

```python
student['grade']=4
student
```

```python
#loop dictionary
people = [
    {"name": "Alice", "id": 1, "age": 25},
    {"name": "Bob", "id": 2, "age": 30},
    {"name": "Charlie", "id": 3, "age": 18}
]
```

```python
for person in people:
    print(person)
```

```python
for person in people:
    print(f"{person['name']} has ID {person['id']} and is {person['age']} years old")
```

```python
for person in people:
    print("Person info:")
    for key, value in person.items():
        print(f"{key}: {value}")
    print("-----")
```
