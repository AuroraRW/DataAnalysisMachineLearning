## Mutable vs Immutable

### Immutable

```python
# int, string, float, bool, tuple
# Create a new objet, when change the value
x=2
x=3
```

```python
# But small ints and string saved in cashe
x = 256
y = 256
print(x is y)   # small integer saved in cashe, -5 to 256
```

### Mutable

```python
# list, dict, set
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b) # same value
print(a is b) # different list saved in different memory location
```

```python
a = [1, 2, 3]
c = a

print(c==a)
print(c is a) # like pointer to assign the address

c.append(4)
print(a)  # [1, 2, 3, 4]
```

```python
# Copy from a list
a = [1, 2, 3]
# c = list(a)
c = a.copy() #shallow copy

print(c is a)   
print(c == a) 

c.append(4)

print(a)  # [1, 2, 3] 
print(c)  # [1, 2, 3, 4]
```

```python
import copy
a = [1, [2, [3,4]]]
# c = list(a)  # Not a deep copy !
c = copy.deepcopy(a)

print(c is a) 
print(c == a)

c[1][1].append(4)

print(a)  # [1, [2, 3, 4]] 
print(c)  # [1, [2, 3, 4]]
```
