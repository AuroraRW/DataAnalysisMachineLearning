## Build-in Function

### Zip

```python
a = (5, 7, 12)
b = (0, -1, 34)

x = zip(a, b)
list(x)
```

### Map

```python
def myfunc(a):
  return len(a)

x = map(myfunc, ['apple', 'banana', 'cherry'])

list(x)
```

```python
def myfunc(a, b):
  return a + b

x = map(myfunc, ['apple', 'banana', 'cherry'], ['orange', 'lemon', 'pineapple'])

list(x)
```

```python
x = map(lambda a, b: a + b,
        ['apple', 'banana', 'cherry'],
        ['orange', 'lemon', 'pineapple'])

list(x)
```

### Filter

```python
ages = [5, 12, 17, 18, 24, 32]

def myFunc(x):
  if x < 18:
    return False
  else:
    return True

adults = filter(myFunc, ages)
list(adults)
```

```python
ages = [5, 12, 17, 18, 24, 32]

adults = filter(lambda x: x >= 18, ages)

list(adults)
```

```python
# Create a list of tuples (x, y) where x is from 1 to 3, y is from 1 to 2. Output: [(1, 1), (1, 2), (2, 1), (2, 2), (3, 1), (3, 2)]
[(x, y) for x in range(1, 4) for y in range(1, 3)]
```

```python
# Given nums = [1, -2, 3, -4, 5], create a list of absolute values only for negative numbers.
nums=[1, -2, 3, -4, 5]
[abs(num) for num in nums if num < 0]
```

```python
# use lambda, filter, map to rewrite
negatives = filter(lambda x: x < 0, nums)
result = list(map(lambda x: abs(x), negatives))
result
```
